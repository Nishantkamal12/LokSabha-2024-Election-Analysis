```python
import pandas as pd
import matplotlib.pyplot as plt

# LOAD DATASET
df = pd.read_csv("/content/GE_2024_Results new.csv")

# CHECK DATA
print(df.head())

# CHECK COLUMNS
print(df.columns)

# PARTY GROUPING
def classify_party(party):

    if party == 'Bharatiya Janata Party':
        return 'BJP'

    elif party == 'Indian National Congress':
        return 'Congress'

    else:
        return 'Others'

# CREATE NEW COLUMN
df['Party_Group'] = df['Party'].apply(classify_party)

# COLORS
bjp_color = '#FF9933'

congress_color = '#0000FF'

others_color = '#D3D3D3'
```

```python
#BJP vs Congress vs Others Total Votes

vote_share = df.groupby(
'Party_Group'
)['Total Votes'].sum()

colors = [
bjp_color,
congress_color,
others_color
]

vote_share.plot(

kind='bar',

figsize=(10,6),

color=colors,

edgecolor='black',

linewidth=2

)

plt.title(
'BJP vs Congress vs Others Total Votes',
fontsize=18,
fontweight='bold'
)

plt.xlabel(
'Party Group',
fontsize=14
)

plt.ylabel(
'Total Votes',
fontsize=14
)

plt.xticks(rotation=0)

plt.grid(axis='y')

plt.tight_layout()

plt.savefig(
'graph1_total_votes.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```

```python
#Seat Share Comparison

seats = df[
df['Result'] == 'Won'
]

seats = seats.groupby(
'Party_Group'
).size()

colors = [
bjp_color,
congress_color,
others_color
]

seats.plot(

kind='pie',

figsize=(8,8),

autopct='%1.1f%%',

colors=colors,

wedgeprops={
'edgecolor':'black',
'linewidth':2
}

)

plt.title(
'Seat Share Comparison',
fontsize=18,
fontweight='bold'
)

plt.ylabel('')

plt.tight_layout()

plt.savefig(
'graph2_seat_share.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```

```python
#BJP Stronghold States

bjp_states = df[

(df['Party'] == 'Bharatiya Janata Party')

&

(df['Result'] == 'Won')

]

bjp_states = bjp_states.groupby(
'State'
).size()

bjp_states = bjp_states.sort_values(
ascending=False
).head(10)

bjp_states.plot(

kind='barh',

figsize=(12,6),

color=bjp_color,

edgecolor='black',

linewidth=2

)

plt.title(
'Top BJP Stronghold States',
fontsize=18,
fontweight='bold'
)

plt.xlabel(
'Seats Won',
fontsize=14
)

plt.ylabel(
'State',
fontsize=14
)

plt.grid(axis='x')

plt.tight_layout()

plt.savefig(
'graph3_bjp_states.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```

```python
#Congress Stronghold States

cong_states = df[

(df['Party'] == 'Indian National Congress')

&

(df['Result'] == 'Won')

]

cong_states = cong_states.groupby(
'State'
).size()

cong_states = cong_states.sort_values(
ascending=False
).head(10)

cong_states.plot(

kind='barh',

figsize=(12,6),

color=congress_color,

edgecolor='black',

linewidth=2

)

plt.title(
'Top Congress Stronghold States',
fontsize=18,
fontweight='bold'
)

plt.xlabel(
'Seats Won',
fontsize=14
)

plt.ylabel(
'State',
fontsize=14
)

plt.grid(axis='x')

plt.tight_layout()

plt.savefig(
'graph4_congress_states.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```

```python
#Top 10 Candidates by Total Votes

top_candidates = df.sort_values(
'Total Votes',
ascending=False
).head(10)

candidate_colors = []

for party in top_candidates['Party']:

    if party == 'Bharatiya Janata Party':
        candidate_colors.append(bjp_color)

    elif party == 'Indian National Congress':
        candidate_colors.append(congress_color)

    else:
        candidate_colors.append(others_color)

plt.figure(figsize=(12,7))

plt.barh(

top_candidates['Candidate'],

top_candidates['Total Votes'],

color=candidate_colors,

edgecolor='black'

)

plt.title(
'Top 10 Candidates by Total Votes',
fontsize=18,
fontweight='bold'
)

plt.xlabel(
'Total Votes',
fontsize=14
)

plt.ylabel(
'Candidate',
fontsize=14
)

plt.grid(axis='x')

plt.tight_layout()

plt.savefig(
'graph7_top_candidates_total_votes.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```

```python
#Bottom 10 Winners by Lowest Victory Margin

# WINNING CANDIDATES
winners = df[df['Result'] == 'Won']

# RUNNER UP VOTES
runner_up = df.groupby(
['State','Constituency']
)['Total Votes'].nlargest(2).reset_index()

runner_up = runner_up.groupby(
['State','Constituency']
).nth(1).reset_index()

runner_up = runner_up.rename(
columns={
'Total Votes':'Runner_Up_Votes'
}
)

# MERGE DATA
margin_df = pd.merge(

winners,

runner_up,

on=['State','Constituency']

)

# VICTORY MARGIN
margin_df['Victory_Margin'] = (

margin_df['Total Votes']

-

margin_df['Runner_Up_Votes']

)

# LOWEST 10 WINNERS
bottom_10 = margin_df.sort_values(
'Victory_Margin'
).head(10)

# COLORS
candidate_colors = []

for party in bottom_10['Party']:

    if party == 'Bharatiya Janata Party':
        candidate_colors.append(bjp_color)

    elif party == 'Indian National Congress':
        candidate_colors.append(congress_color)

    else:
        candidate_colors.append(others_color)

# GRAPH
plt.figure(figsize=(12,7))

plt.barh(

bottom_10['Candidate'],

bottom_10['Victory_Margin'],

color=candidate_colors,

edgecolor='black'

)

# TITLE
plt.title(
'Bottom 10 Winners by Lowest Victory Margin',
fontsize=18,
fontweight='bold'
)

# LABELS
plt.xlabel(
'Victory Margin',
fontsize=14
)

plt.ylabel(
'Candidate',
fontsize=14
)

# GRID
plt.grid(axis='x')

# SAVE
plt.tight_layout()

plt.savefig(
'graph11_bottom10_lowest_margin.png',
dpi=300,
bbox_inches='tight'
)

# SHOW
plt.show()
```

```python
#State Wise Political Control
winner_df = df[
df['Result'] == 'Won'
]

state_party = winner_df.pivot_table(

index='State',

columns='Party_Group',

values='Candidate',

aggfunc='count',

fill_value=0

)

state_party = state_party.sort_values(
by='BJP',
ascending=False
).head(10)

state_party.plot(

kind='bar',

stacked=True,

figsize=(14,7),

color=[
bjp_color,
congress_color,
others_color
],

edgecolor='black'

)

plt.title(
'State Wise Political Control',
fontsize=18,
fontweight='bold'
)

plt.xlabel(
'State',
fontsize=14
)

plt.ylabel(
'Seats Won',
fontsize=14
)

plt.xticks(rotation=45)

plt.grid(axis='y')

plt.tight_layout()

plt.savefig(
'graph9_state_political_control.png',
dpi=300,
bbox_inches='tight'
)

plt.show()
```
